```python
# This program reads in a specified CSV data table of galaxies and plots multiple histograms of the sample data
# I will also use this program to select sample objects from the individual redshift bins 

## IMPORTS ## 

import matplotlib                        # for plotting
import matplotlib.pyplot as plt          # shortcut for plot command
import matplotlib.lines as mplines       # lines for plotting
import numpy as np                       # 'arange' function to define steps list
import csv                               # to read in CSV file
import cosmolopy.distance as cd
import math
import random



## DEFINE NEW FUNCTIONS HERE ##

def absmag(zbin):    
    """Calulates the absolute magnitude of a set of apparent magnitudes"""
    # Arguments
    # zbin - 2D list containing redshifts in column 13
    
    rabs = []
    for row in zbin:
        lumd0=cd.luminosity_distance(row[col['z']], **cosmo) * 10**6
        Mr = row[col['r']] + 5. - 5. * math.log10(lumd0)
        row.append(Mr)
#        rabs.append(row[col['Mr']])
        rabs.append(Mr)
    return rabs, zbin


def sgauss(N, mu, sigma):
    """Selects a set of N numbers from a Gaussian distribution of average mu and width sigma"""
    result = []
    for i in range(N):
        result.append(random.gauss(mu, sigma))
    return result


def matchobj(select, binap):
    """Select sample objects with given magnitudes from the objects in the bin"""
    # Function arguments #
    # select - set of absolute magnitudes to select from binap
    # binap - full table of the bin

    values = []      # 2D list of possible absolute magnitudes of objects to be selected
    objects = []     # 2D list to store the objects' rows
    
    for absmag in select:
        tempbin = []
        for row in binap:
            # Apply the observability cuts for Semester B from Gemini South
                if row[col['dec']] < 15.0 and (row[col['ra']] > 300. or row[col['ra']] < 120.):
                    tempbin.append(abs(row[col['Mr']] - absmag))
        tempbin.sort()
        values.append([tempbin[0]+absmag, absmag-tempbin[0]])        
    
    for mags in values:
        for row in binap:
            if row[col['Mr']] == mags[0]: objects.append(row)
            elif row[col['Mr']] == mags[1]: objects.append(row)
    
    return objects

```

```python
## MAIN ##

cosmo = {'omega_M_0' : 0.3, 'omega_lambda_0' : 0.7, 'h' : 0.72}
cosmo = cd.set_omega_k_0(cosmo)

redshifts = []        # Define an unpopulated list to store the redshift values
gband = []            # Define an unpopulated list to store the g band magnitudes
rband = []            # Define an unpopulated list to store the r band magnitudes
bin0 = []             # Define an unpopulated two dimensional list for data table of objects
bin1 = []
bin2 = []
bin3 = []

total = []            # Define an empty list to store the full csv file as 2D list


# Create a dictionary with the CSV file information
# CSV file contains the following:
# specObjID,ra,dec,Hbeta,sigma,OIII,sigma,Ca K,CaK_EW,Ca H,CaH_EW,g,r,z

col = {'specobjid':0, 'ra':1, 'dec':2, 'Hbeta':3, 'Hbsig':4, 'OIII':5, 'OIIIsig':6, 'CaK':7, 'CaKew':8, 
         'CaH':9, 'CaHew':10, 'g':11, 'r':12, 'z':13, 'Mr':14}


# Read in the CSV file, parse the current row, and store the last item from the row in the redhsifts list

with open('NLS1_final_noimg_khiner_mod.csv') as csvfile:
    infor = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC)        
    #QUOTE_NONNUMERIC indicates the values are to be read as floats
    for row in infor:
        total.append(row)
        redshifts.append(row[col['z']])
        gband.append(row[col['g']])
        rband.append(row[col['r']])
        if row[col['z']] <= 0.1:        # This reads in the object fields for all obj z < 0.1
            bin0.append(row)
        if row[col['z']] > 0.1 and row[col['z']] <= 0.2:        
            bin1.append(row)
        if row[col['z']] > 0.2 and row[col['z']] <= 0.3:        
            bin2.append(row)
        if row[col['z']] > 0.3 and row[col['z']] <= 0.4:        
            bin3.append(row)            
            
        
# Test if the file is closed
if csvfile.closed: print 'CSV file is closed'
      

        
# Calculate the g-r colors
color = [x - y for x, y in zip(gband, rband)]

# Define the bin sizes for the histogram
zsteps = np.arange(0.,1.,0.05)        #redshifts


# print bin0 before calculating absolute magnitudes
print "First object of bin0 ", bin0[0], "\n"

# Calculate the absolute magnitudes 
# Note this does not (yet) contain a K-correction
rabs0, bin0ap = absmag(bin0)     # Calls the function absmag that I defined at the top
rabs1, bin1ap = absmag(bin1)
rabs2, bin2ap = absmag(bin2)
rabs3, bin3ap = absmag(bin3)
totabs, totap = absmag(total)

rabs = [rabs0, rabs1, rabs2, rabs3]

# Test some output
print "Bin0ap, first object ", bin0ap[0], "\n"


#rdev0 = np.std(rband0)

# Calculate the average and standard deviation of the absolute magnitudes. 
rabsdev = [np.std(rabs0), np.std(rabs1), np.std(rabs2), np.std(rabs3)]
rabsav = [np.mean(rabs0), np.mean(rabs1), np.mean(rabs2), np.mean(rabs3)]
totabsdev = np.std(totabs)
totabsav = np.std(totabs)

# Print some info on the bin stats
#print "The average and standard deviation of the r band mag in Bin0 is", rdev0
print "The average and standard deviation of the r band absolute mag in Bin0 is", rabsav[0], rabsdev[0]
print "The average and standard deviation of the r band absolute mag in Bin1 is", rabsav[1], rabsdev[1]
print "The average and standard deviation of the r band absolute mag in Bin2 is", rabsav[2], rabsdev[2]
print "The average and standard deviation of the r band absolute mag in Bin3 is", rabsav[3], rabsdev[3], "\n"

 
# Now select sample based on the distribution of the absolute magnitudes in each redshift bin

sample0 = sgauss(8, rabsav[0], rabsdev[0])     # Select magnitudes from a Gaussian distribution
print "Luminosities selected from zeroth redshift bin", sample0, "\n"
objlist0 = matchobj(sample0, bin0ap)     # Match the Magnitudes to the sample 
#print "Test run gives test values", test0, "\n"

print "The following objects have been selected for Bin0"
for row in objlist0: print row, "\n"
    
sample1 = sgauss(8, rabsav[1], rabsdev[1])
objlist1 = matchobj(sample1, bin1ap)

print "Objects have been selected for Bin1"

sample2 = sgauss(8, rabsav[2], rabsdev[2])
objlist2 = matchobj(sample2, bin2ap)

print "Objects have been selected for Bin2"

sample3 = sgauss(8, rabsav[3], rabsdev[3])
objlist3 = matchobj(sample3, bin3ap)

print "Objects have been selected for Bin3 \n"



# Print the selected objects to an ouput file 

sampfile = open('junk', 'w')

sampfile.write("Selected Sample for Gemini Fall Semester\n  20h < RA < 08h, DEC < 15\n")
sampfile.write("z <= 0.1 \n")
for row in objlist0: 
    sampfile.write(str(row))
    sampfile.write("\n")
sampfile.write("\n")
sampfile.write("0.1 < z <= 0.2 \n")
for row in objlist1: 
    sampfile.write(str(row))
    sampfile.write("\n")
sampfile.write("\n")
sampfile.write("0.2 < z <= 0.3 \n")
for row in objlist2: 
    sampfile.write(str(row))
    sampfile.write("\n")
sampfile.write("\n")
sampfile.write("0.3 < z <= 0.4 \n")
for row in objlist3: 
    sampfile.write(str(row))
    sampfile.write("\n")

sampfile.close()
print "Wrote selected sample to geminisample"
if sampfile.closed: print "File geminisample is closed"
    
    
# Print objects from redshifts 0.2 - 0.4 into a file with the following
# Name,RA,DEC

#with open('gsaoibins', 'w') as csvfile: 
#    spamwriter = csv.writer(csvfile, quoting=csv.QUOTE_MINIMAL)
#    count = 0
#    for row in bin2:
#        spamwriter.writerow('SDSS'+str(count))
#        spamwriter.writerow(str(row[col['ra']])+str(row([col['dec']]))
   
#print "gsaoibins is closed"


gsaoi = open('junk2', 'w')
count = 0
#gsaoi.write("Name,RAJ2000,DecJ2000,pmRA,pmDec,U,U_sys,B,B_sys,V,V_sys,UC,UC_sys,R,R_sys,I,I_sys,Y,Y_sys,J,J_sys,H,H_sys,Kmag,K_sys,L,L_sys,M,M_sys,N,N_sys,Q,Q_sys
#)
gsaoi.write("Name,RAJ2000,DecJ2000,r,z,V\n")


for row in bin2:
    if row[col['dec']] < 15.0 and (row[col['ra']] > 300. or row[col['ra']] < 120.):    
        gsaoi.write('SDSS'+str(count) + ',')
        gsaoi.write(str(row[col['ra']]) + ',')
        gsaoi.write(str(row[col['dec']]) + ',')
        gsaoi.write(str(row[col['r']]) + ',')
        gsaoi.write(str(row[col['z']]) + ',')
        V = row[col['g']] - 0.52 * (row[col['g']] - row[col['r']]) - 0.03
        gsaoi.write(str(V) + '\n')
        count+=1
    
for row in bin3:
    if row[col['dec']] < 15.0 and (row[col['ra']] > 300. or row[col['ra']] < 120.):
        gsaoi.write('SDSS'+str(count) + ',')
        gsaoi.write(str(row[col['ra']]) + ',')
        gsaoi.write(str(row[col['dec']]) + ',')
        gsaoi.write(str(row[col['r']]) + ',')
        gsaoi.write(str(row[col['z']]) + ',')
        V = row[col['g']] - 0.52 * (row[col['g']] - row[col['r']]) - 0.03
        gsaoi.write(str(V) + '\n')
        count+=1
    
    
gsaoi.close()

print "gsaoibins is closed"
```
