# Display Spectra 

This notebook will create plots of the spectra that I just reduced. I will need to read in a fits file and plot a spectrum

```python
from astropy.io import fits
```

```python
# Read and plot function
def rdnorm_spec(filename, z, limits):
    '''Read in a spectrum given the full path filename and return the rest frame
       wavelength array and the normalized flux array

    INPUT
        filename  -  full path of spectrum to plot
        z         -  object redshift
        limits    -  wavelength limits for normalization in [x1, x2]

    OUTPUT 
        lam       -  wavelength array in the rest frame
        spec_data -  flux array normalized within limits'''
    
    spectrum = fits.open(filename)
    
    # Define the wavelength array
    lam0 = float(spectrum[0].header['CRVAL1'])
    dlam = float(spectrum[0].header['CD1_1'])
    length = int(spectrum[0].header['NAXIS1'])
    lam = lam0 + dlam*np.array(range(0,length))
    lam = lam / (1. + z)
    
    # Define the data array 
    spec_data = spectrum[0].data
    
#    spec_data = spec_data - spec_data[np.where((lam > limits[0]) & (lam < limits[1]))].mean()
    
    spectrum.close
    
    return lam, spec_data
    
def plt_spec(objname, lam_arr, flux_arr, z, limits, count, include):
    '''Plot the spectrum within the defined wavelength limits

    INPUTS
        objname   -  object name
        lam_arr   -  wavelength array
        flux_arr  -  flux array
        z         -  redshift of object
        limits    -  range of wavelengths to plot
        count     -  the number of times the plot has been called - used to create flux offset
        include   -  list of explicit objects to include

    OUTPUTS
        count     -  an updated number of times the plot has been made'''
    
    # Plot the data and close the file
    
    if objname in include:
        figure()
        plot(lam_arr, flux_arr, '-g', linewidth=2)# + (count*2000.))
        count += 1
    
        xlim(limits[0], limits[1])
        ylim(0,)
        if objname == 'SDSS 233832.8-092059': ylim(0, 6000)
        if objname == 'SDSS 225730.6-002033': ylim(0, 6000)
        title(objname + ', z = ' + str(z), fontsize = 18)
        xlabel('Rest-frame Wavelength (A)', fontsize = 16)
        ylabel('Counts', fontsize=16)
    
    return count
```
